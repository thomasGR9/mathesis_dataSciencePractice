```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
natsal = pd.read_csv('../natsal_het_sex_partners.csv')
natsal
```

```python
natsal.groupby('Sex').describe()
```

```python
natsal.groupby('Sex').agg(lambda x: x.value_counts().index[0])
```

```python
sns.set_style('darkgrid')
sns.displot(natsal, x='NumPartners')
plt.show()
```

```python
natsal_50 = natsal[natsal['NumPartners'] <= 50]
sns.displot(natsal_50, x='NumPartners')
plt.show()
```

```python
sns.displot(natsal_50, x='NumPartners', height=7, aspect=1.5)
plt.show()
```

```python
sns.displot(natsal_50, x='NumPartners',hue='Sex' ,height=7, aspect=1.5)
plt.show()
```

```python
sns.displot(natsal_50, x='NumPartners',hue='Sex',palette={'M': 'blue', 'F':'pink'}, height=7, aspect=1.5)
plt.show()
```

```python
sns.displot(natsal_50, x='NumPartners',hue='Sex',palette={'M': 'blue', 'F':'pink'},multiple='dodge' ,height=7, aspect=1.5)
plt.show()
```

```python
sns.displot(natsal_50, x='NumPartners',hue='Sex',edgecolor='black',palette={'M': 'blue', 'F':'pink'},multiple='dodge' ,height=7, aspect=1.5)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=natsal_50['NumPartners'],y=natsal_50['Sex'],palette={'M': 'blue', 'F':'pink'})
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.stripplot(x=natsal_50['NumPartners'],y=natsal_50['Sex'],edgecolor='black',palette={'M': 'blue', 'F':'pink'}, size=7, linewidth=0.5,jitter=0.25)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.violinplot(x=natsal_50['NumPartners'],y=natsal_50['Sex'],edgecolor='black',palette={'M': 'blue', 'F':'pink'})
plt.show()
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', 25)
gdp_df = pd.read_csv('../worldbank/gdp_per_capita_ppp/API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3469401.csv', skiprows=4)
```

```python
gdp_df
```

```python
gdp_df.drop(gdp_df.columns[len(gdp_df.columns) - 1], inplace=True, axis=1)
```

```python
gdp_df
```

```python
countries_df = pd.read_csv('../worldbank/gdp_per_capita_ppp/Metadata_Country_API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3469401.csv')
countries_df
```

```python
 countries_df.drop(countries_df.columns[len(countries_df.columns) - 1], inplace=True, axis=1)
countries_df = countries_df.loc[countries_df['Region'].notnull()]
countries_df
```

```python
gdp_countries_df = pd.merge(gdp_df, countries_df[['Country Code', 'Region']], how='inner', on='Country Code')
```

```python
gdp_countries_df
```

```python
lex_df = pd.read_csv('../worldbank/life_expectancy/API_SP.DYN.LE00.IN_DS2_en_csv_v2_3469289.csv', skiprows=4)
```

```python
lex_df
```

```python
lex_df.drop(lex_df.columns[len(lex_df.columns) - 1], inplace=True, axis=1)
```

```python
lex_gdp_countries_df = pd.merge(gdp_countries_df, lex_df.loc[:, 'Country Code':], how='inner', on='Country Code', suffixes=('_gdp', '_lex'))
```

```python
lex_gdp_countries_df
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
```

```python
import matplotlib.patches as patches

plt.style.use('ggplot')
plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])

rect1=patches.Rectangle((0, 52), width=25000, height=30, color='blue', alpha=0.2)
plt.gca().add_patch(rect1)
rect2=patches.Rectangle((25000, 72), width=115000, height=14, color='gold', alpha=0.2)
plt.gca().add_patch(rect2)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
plt.xscale('log')
```

```python
population_df = pd.read_csv('../worldbank/population/API_SP.POP.TOTL_DS2_en_csv_v2_3603750.csv', skiprows=4)
population_df.drop(population_df.columns[len(population_df.columns) - 1], inplace=True, axis=1)
```

```python
population_df
```

```python
pop_lex_gdp_countries_df = pd.merge(lex_gdp_countries_df, population_df, how='inner', on='Country Code')
```

```python
pop_lex_gdp_countries_df
```

```python
pop_column_names = { c: f' {c}_pop' for c in population_df.columns if '1960' <= c <= '2020'}
pop_lex_gdp_countries_df.rename(columns=pop_column_names, inplace=True)
```

```python
pop_lex_gdp_countries_df
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x = pop_lex_gdp_countries_df['2019_gdp'], y = pop_lex_gdp_countries_df['2019_lex'], alpha=0.5, s=pop_lex_gdp_countries_df[' 2019_pop']/10**6)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x = pop_lex_gdp_countries_df['2019_gdp'], y = pop_lex_gdp_countries_df['2019_lex'], alpha=0.5, s=pop_lex_gdp_countries_df[' 2019_pop']/10**6)
plt.xscale('log')
```

```python
region_cat = pop_lex_gdp_countries_df['Region'].sort_values().astype('category')
pop_lex_gdp_countries_df['Region Code'] = region_cat.cat.codes
```

```python
region_cat.cat.categories
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10, 8))
scatter = plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'],
                      y=pop_lex_gdp_countries_df['2019_lex'],
                      s=pop_lex_gdp_countries_df[' 2019_pop']/10**6,
                      c=pop_lex_gdp_countries_df['Region Code'],
                      alpha=0.6,
                      edgecolors='k',
                      label=pop_lex_gdp_countries_df['Region'],
                      cmap='Spectral_r')
plt.xscale('log')
handles, labels = scatter.legend_elements(prop='colors', num=len(region_cat.cat.categories), size=10, mec='k', alpha=0.6)
labels = region_cat.cat.categories
plt.legend(handles,labels ,loc='lower right')
```

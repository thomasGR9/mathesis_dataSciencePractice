```python

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Simple plot')

plt.legend()
```

```python
fig = plt.figure()

axes = fig.gca()
axes.plot(x, x, label='linear')

axes.plot(x, x**2, label='quadratic')
axes.plot(x, x**3, label='cubic')

axes.set_xlabel('x label')
axes.set_ylabel('y label')

axes.set_title('Simple plot')

axes.legend()
```

```python
legend_loc_str = [
    'best',
    'upper right',
    'upper left',
    'lower left',
    'lower right',
    'right',
    'center left',
    'center right',
    'lower center',
    'upper center',
    'center'
]

plt.style.use('ggplot')
plt.figure(figsize=(12, 24))
x = np.linspace(0, 2, 100)
plt.subplots_adjust(hspace=0.4)

for i, loc in enumerate(legend_loc_str):
    plt.subplot(6, 2, i + 1)
    plt.plot(x, x, label='lineal')
    plt.plot(x, x**2, label='quadratic')
    plt.plot(x, x**3, label='cubic')

    plt.xlabel('x label')
    plt.ylabel('y label')
    plt.title('Simple plot, legend at = ' + loc)
    plt.legend(loc=loc)
```

```python
figure = plt.figure(figsize=(12, 24))
figure.subplots_adjust(hspace=0.4)

x = np.linspace(0, 2, 100)

for i, loc in enumerate(legend_loc_str):
    ax = figure.add_subplot(6, 2, i + 1)
    ax.plot(x, x, label='linear')
    ax.plot(x, x**2, label='quadratic')
    ax.plot(x, x**3, label='cubic')

    ax.set_xlabel('x label')
    ax.set_ylabel('y label')

    ax.set_title('Simple plot, legend at =' + loc)

    ax.legend(loc=loc)
```

```python
colors = [
     'b',
    'g',
    'r',
    'c',
    'm',
    'y',
    'k',
    'w'
]
figure = plt.figure(figsize=(12, 24))
figure.subplots_adjust(hspace=0.4)

x = np.linspace(-1, 1, 100)
y = x * np.sin(1 / x)

for i, color in enumerate(colors):
    ax = figure.add_subplot(2, 4 , i + 1)
    ax.plot(x, y, color=color, label='$y = x / sin(1/x)$')
    

    ax.set_xlabel('x label')
    ax.set_ylabel('y label')

    ax.set_title('Simple plot, color set at ' + color)

    ax.legend(loc='best')
```

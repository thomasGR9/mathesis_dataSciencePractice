```python
def lng_lat_to_pixels(lng, lat):    
    lng_rad = lng * np.pi / 180
    lat_rad = lat * np.pi / 180
    x = (256/(2*np.pi))*(lng_rad + np.pi)
    y = (256/(2*np.pi))*(np.log(np.tan(np.pi/4 + lat_rad/2)))
    return (x, y)
```

```python
px, py  = lng_lat_to_pixels(data['pickup_longitude'], data['pickup_latitude'])
px.head(10)
```

```python
 plt.scatter(px, py)
```

```python
plt.figure(figsize=(12, 10))
plt.axis('equal')
plt.xlim(75.35, 75.5)
plt.ylim(31.6, 31.95)
plt.axis('off')
plt.gca().set_facecolor('white')
plt.scatter(px, py, s=0.1, alpha=.03, color='black')
```

```python
trip_distance_counts = data['trip_distance'].value_counts(bins=np.linspace(0, 10 , 100)).sort_index()
trip_distance_counts
```

```python
from scipy.signal import find_peaks
peaks, _ = find_peaks(trip_distance_counts)
peaks
```

```python
trip_distance_counts.iloc[peaks]
```

```python
distance_hist = data.trip_distance.hist(bins=np.linspace(0, 10, 100), color='blue', edgecolor='black', alpha=0.5)

for i in peaks:
    distance_hist.patches[i].set_facecolor('red')
```
